```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_bior = pd.read_csv('./runs/detect/bior/results.csv')
df_bior.columns = ['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss', "", "", ""]
df_bior = df_bior[['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss']]
df_bior = df_bior.iloc[:150, :]
df_bior
```

```python
df = pd.read_csv('./runs/detect/normal/results.csv')
df.columns = ['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss', "", "", ""]
df = df[['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss']]
df
```

```python
df_haar = pd.read_csv('./runs/detect/haar/results.csv')
df_haar.columns = ['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss', "", "", ""]
df_haar = df_haar[['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss']]
df_haar
```

```python
plt.figure(figsize=(12, 8))

# Plotting with different line styles, markers, and colors
plt.plot(df[["train/cls_loss"]], label="No model", linestyle='-', color="purple")
plt.plot(df_bior[["train/cls_loss"]], label="bior1.3", linestyle='-', color="blue")
plt.plot(df_haar[["train/cls_loss"]], label="haar", linestyle='-', color="green")

# Adding grid and adjusting legend
plt.grid(True)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Categorical Cross-Entropy Loss")
plt.legend(loc="upper right")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(df[["metrics/precision(B)"]], label="No Model", linestyle='-', color="purple")
plt.plot(df_bior[["metrics/precision(B)"]], label="bior1.3", linestyle='-', color="blue")
plt.plot(df_haar[["metrics/precision(B)"]], label="haar", linestyle='-', color="green")


plt.grid(True)
plt.xlabel("Epochs")
plt.ylabel("Percentual")
plt.title("Precision")
plt.legend(loc="lower right")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(df[["metrics/recall(B)"]], label="No Model")
plt.plot(df_bior[["metrics/recall(B)"]], label="bior1.3")
plt.plot(df_haar[["metrics/recall(B)"]], label="haar", color="green")
plt.xlabel("Epochs")
plt.ylabel("Percentual")
plt.title("recall")
plt.legend()

plt.show()


plt.figure(figsize=(12,8))

plt.plot(df[["metrics/recall(B)"]], label="No Model", linestyle='-', color="purple")
plt.plot(df_bior[["metrics/recall(B)"]], label="bior1.3", linestyle='-', color="blue")
plt.plot(df_haar[["metrics/recall(B)"]], label="haar", linestyle='-', color="green")


plt.grid(True)
plt.xlabel("Epochs")
plt.ylabel("Percentual")
plt.title("Recall")
plt.legend(loc="upper right")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(df[["metrics/mAP50(B)"]], label="No Model")
plt.plot(df_bior[["metrics/mAP50(B)"]], label="bior1.3")
plt.plot(df_haar[["metrics/mAP50(B)"]], label="haar")
plt.xlabel("Epochs")
plt.ylabel("Percentual")
plt.title("mAP50")
plt.legend()

plt.show()
```
